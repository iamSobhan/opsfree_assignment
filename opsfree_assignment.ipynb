<p>a. What algorithm, machine learning or AI approaches would you take to find anomalies in the duration of a span? And why do you think that approach is a good approach?</p>


<b>Answer</b>
<p>The given problem assignment is a Unsupervised Machine Learning problem.<br>
Unsupervised Machine Learning algorithm is a Machine Learning algorithm, that is used to find  patterns on the set of given data.<br>
                                                                                                      
There are a few Machine Learning and Artificial Intelligence methods for identifying abnormalities fora span,</p>

<b>Autoencoders</b>
<p>An autoencoder is a kind of neural network where the input and output layers must have the same dimensions. By comparing the input and output, reconstruction errors caused by anomalies can be identified.<br>
Autoencoders are advantageous because,<br>
(i.) Complex nonlinear relationships in the data can be captured by them.<br>
(ii.) They canpick up data representations, which could increase detection precision.<br>
(iii.) They are capable of handling several kinds of data, including time seriesand other sequential data.</p>

<b>Statistical Z-Score Method</b>
<p>Determine the Z-score for every dataset duration value. The Z-score calculates a value's departure from the mean in standard deviations.<br>
High absolute Z-score values can be regarded as outliers. When dealing with regularly distributed data, this approach is straightforward and useful for locating outlier spots.</p>
    
<b>One-Class SVM</b>
<p>One-Class SVM is a type of Support Vector Machines designed for outlier detection. It learns a decision boundary around the normal data points, effectively identifying anomalies as points lying outside this boundary. <br>
This method is advantageous because,<br>
(i.) It can capture complex relationships in the data using kernel tricks.<br>
(ii.) It can handle high-dimensional data efficiently.<br>
(iii.) It's effective even with limited training data.</p>

<p>These techniques can be used to various datasets and dimensions and are capable of handling a wide range of data distributions. These methods are effective in a variety of situations, including those with irregular distributions, high-dimensional data, and time-series data. Particularly for large datasets, several of these methods are effective and fast in identifying anomalies.<br>
Overall, These methods can handle high-dimensional data, are resilient to outliers, and don't require labeledanomalies for training, making them appropriate for identifying anomalies in the length of spans overall. They are very flexible for tasks involving anomaly detection since they can adjust to various data distributions and capture intricate correlations.</p>
